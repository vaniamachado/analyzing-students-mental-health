```python
# ==============================
# Students Mental Health Project
# ==============================

# 1️⃣ Install dependencies (run once)
!pip install sqlalchemy psycopg2-binary pandas --quiet

# 2️⃣ Import packages
import pandas as pd
from sqlalchemy import create_engine

# 3️⃣ Create SQLAlchemy engine
# Replace username/password/dbname with your PostgreSQL setup
engine = create_engine('postgresql://postgres:sIlv3stris@localhost:2309/students')

# 4️⃣ Query: Aggregate mental health scores for international students by length of stay
query = """
SELECT stay,
       COUNT(*) AS count_int,
       ROUND(AVG(todep),2) AS average_phq,
       ROUND(AVG(tosc),2) AS average_scs,
       ROUND(AVG(toas),2) AS average_as
FROM students_mental_health
WHERE inter_dom = 'Inter'
GROUP BY stay
ORDER BY stay DESC;
"""

# 5️⃣ Execute query and load into DataFrame
df_result = pd.read_sql(query, engine)

# 6️⃣ Display results
print("=== International Students Mental Health Summary ===")
df_result
```
